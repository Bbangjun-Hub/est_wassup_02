# 사용 컬럼
+ 평균 해면기압 - 하루 동안 관측된 해면기압의 평균값  
+ 합계 일사량 - 일조시간 / 가조시간   
+ 평균 지면온도 - 일정조건에서 측정된 지면온도   
+ 최저 초상온도 - 지표에 접해 있는 풀 위의 공기온도, 서리 발생을 가늠하는 척도   
+ 평균 풍속 - 특정 기간 동안의 평균 풍속을 미터/초 단위로 나타낸 것   
+ 최다 풍향 - 특정 기간 동안 가장 자주 나타나는 풍향을 16방위로 표현한 것   
+ 평균 상대습도 - 특정 기간 동안의 평균 상대습도로, 공기에 포함된 수증기의 평균 양에 대한 백분율   
+ 일 최심적설 - 눈이 내린 후, 그 중에서 가장 두꺼운 층의 두께   
+ 평균 전운량  - 하루동안 구름 양을 나타내는 지표 10으로 갈수록 구름이 많아짐  
+ 평균기온(°C)  
+ 최저기온(°C)  
+ 최고기온(°C)  
+ 강수 계속시간(hr)  
+ 일강수량(mm)

```python
no2_null = pd.read_csv("/home/dataset/no2_null_values.csv")
no2_null.info()
```

```python
no2_null.rename(columns={"날짜" : "일시"}, inplace=True)
no2_null["일시"] = pd.to_datetime(no2_null["일시"])
no2_null
```

```python
merge_df = pd.merge(merge_df, no2_null, on= ["일시"], how= "left")
mask = (merge_df['일시'].dt.year == 2021) & (merge_df['일시'].dt.month == 6)
merge_df.loc[mask, 'no2_x'] = merge_df.loc[mask, 'no2_x'].fillna(merge_df['no2_y'])
merge_df = merge_df.drop("no2_y", axis=1)
merge_df.rename(columns={"no2_x" : "no2"}, inplace=True)
```

```python
merge_df.isnull().sum()
```

```python
merge_df['평균 현지기압(hPa)'] = merge_df['평균 현지기압(hPa)'].fillna(1011.6)
```

```python
merge_df['평균 해면기압(hPa)'] = merge_df['평균 해면기압(hPa)'].fillna(1019.2)
```

```python
merge_df.isnull().sum()
```
