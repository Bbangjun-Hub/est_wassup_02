```python
merge_df["최저기온(°C)"] = merge_df["최저기온(°C)"].fillna(25.1)
```

```python
merge_df.isnull().sum()
```

```python
no2_null = pd.read_csv("dataset/no2_null_values.csv")
no2_null.info()
```

```python
no2_null.rename(columns={"날짜" : "일시"}, inplace=True)
no2_null["일시"] = pd.to_datetime(no2_null["일시"])
no2_null
```

```python
merge_df = pd.merge(merge_df, no2_null, on= ["일시"], how= "left")
merge_df.isnull().sum()
```

```python
mask = (merge_df['일시'].dt.year == 2021) & (merge_df['일시'].dt.month == 6)

merge_df.loc[mask, 'no2_x'] = merge_df.loc[mask, 'no2_x'].fillna(merge_df['no2_y'])
```

```python
merge_df = merge_df.drop("no2_y", axis=1)
merge_df.rename(columns={"no2_x" : "no2"}, inplace=True)
merge_df.isnull().sum()
```

```python
merge_df[merge_df["평균 현지기압(hPa)"].isnull()]["평균 현지기압(hPa)"] = 1011.6
```

```python
merge_df['평균 현지기압(hPa)'] = merge_df['평균 현지기압(hPa)'].fillna(1011.6)
```

```python
merge_df['평균 해면기압(hPa)'] = merge_df['평균 해면기압(hPa)'].fillna(1019.2)
```

```python
merge_df.isnull().sum()
```

```python
merge_df.loc[merge_df['일시'] == '2018-7-28', '최저 초상온도(°C)'] = 22.8 # 2022-08-08
```

```python
merge_df.loc[merge_df['일시'] == '2022-08-08', '최저 초상온도(°C)'] = 24.4
```

```python
merge_df.loc[merge_df['일시'] == '2018-02-15', '합계 일사량(MJ/m2)'] = 14.40
merge_df.loc[merge_df['일시'] == '2019-08-28', '합계 일사량(MJ/m2)'] = 6.80
merge_df.loc[merge_df['일시'] == '2019-08-29', '합계 일사량(MJ/m2)'] = 8.22
merge_df.loc[merge_df['일시'] == '2021-01-07', '합계 일사량(MJ/m2)'] = 6.78
merge_df.loc[merge_df['일시'] == '2022-08-08', '합계 일사량(MJ/m2)'] = 1.07
```
